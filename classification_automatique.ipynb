```python
# Utilisation pratique de l'ACP
```

```python
import numpy as np
```

```python
# Génération de données selon une loi normale tridimensionnelle
```

```python
rndn3d = np.random.randn(500,3)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Affichage du nuage des points
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(rndn3d[:,0], rndn3d[:,1], rndn3d[:,2])
plt.title("Données initiales")
plt.show()
```

```python
# Appliquer l'ACP à ces données
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
pca.fit(rndn3d)
```

```python
print("Pourcentage de variance expliquée : ")
print(pca.explained_variance_ratio_)
print("Composantes principales : ")
print(pca.components_)
```

```python
# Générons un autre tableau de données en utilisant np.random.randn(500,3)
rndn3d = np.random.randn(500,3)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(rndn3d[:,0], rndn3d[:,1], rndn3d[:,2])
plt.title("Données initiales")
plt.show()
```

```python
pca = PCA(n_components=3)
pca.fit(rndn3d)
```

```python
print("Pourcentage de variance expliquée : ")
print(pca.explained_variance_ratio_)
print("Composantes principales : ")
print(pca.components_)
```

```python
# Appliquons maintenant une déformation et une rotation dans l'espace tridimensionnel au nuage des observations de rndn3d
```

```python
# Matrice de déformation
s1 = np.array([[3,0,0],[0,1,0],[0,0,0.2]])
# Matrice de rotation
r1 = np.array([[0.36, 0.48,-0.8],[-0.8, 0.6, 0],[0.48, 0.64, 0.6]])
```

```python
rndef = rndn3d.dot(s1).dot(r1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(rndef[:,0], rndef[:, 1], rndef[:, 2])
plt.title("Données déformées")
plt.show()
```

```python
pca = PCA(n_components=3)
pca.fit(rndef)
```

```python
print("Pourcentage de variance expliquée: ")
print(pca.explained_variance_ratio_)
print("Composantes principales : ")
print(pca.components_)
```
